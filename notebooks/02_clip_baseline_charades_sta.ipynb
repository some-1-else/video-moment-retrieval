# 02. CLIP Baseline on Charades-STA

Этот ноутбук показывает базовый CLIP-run на Charades-STA. Он читает уже сохраненные `metrics.json` и `run_config.json`.


```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path.cwd()
if not (ROOT / 'results').exists():
    ROOT = ROOT.parent

def read_json(path):
    with open(ROOT / path, 'r', encoding='utf-8') as f:
        return json.load(f)

def metrics_table(rows):
    return pd.DataFrame(rows)

metrics = read_json('.agent_memory/results/exploratory/charades_sta_baseline_200/metrics.json')
config = read_json('.agent_memory/results/exploratory/charades_sta_baseline_200/run_config.json')
summary = {**config['config'], **config['charades_sta_smoke'], **metrics}
pd.DataFrame([summary])
```
