# 01. Dataset Setup

This notebook is a lightweight reader-facing check of the local data layout used by the coursework. It does not run CLIP, Moment-DETR, or any heavy experiment. For preparation instructions, see `docs/data_setup.md`.


The final experiments use **Charades-STA** with locally prepared Charades videos. QVHighlights remains a preliminary direction because its raw videos depend on external YouTube availability.


```python
from pathlib import Path
import pandas as pd

ROOT = Path.cwd()
if not (ROOT / 'results').exists():
    ROOT = ROOT.parent

paths = [
    'data/charades_sta/charades_sta_train.txt',
    'data/charades_sta/charades_sta_test.txt',
    'data/raw/charades/videos',
    'docs/data_setup.md',
]
status = []
for rel in paths:
    p = ROOT / rel
    status.append({'path': rel, 'exists': p.exists(), 'type': 'dir' if p.is_dir() else 'file'})
pd.DataFrame(status)
```

```python
def count_lines(rel):
    p = ROOT / rel
    if not p.exists():
        return None
    with p.open('r', encoding='utf-8') as f:
        return sum(1 for _ in f)

pd.DataFrame([
    {'split': 'train', 'path': 'data/charades_sta/charades_sta_train.txt', 'rows': count_lines('data/charades_sta/charades_sta_train.txt')},
    {'split': 'test', 'path': 'data/charades_sta/charades_sta_test.txt', 'rows': count_lines('data/charades_sta/charades_sta_test.txt')},
])
```

```python
video_dir = ROOT / 'data/raw/charades/videos'
video_count = len(list(video_dir.glob('*.mp4'))) if video_dir.exists() else 0
pd.DataFrame([{'video_dir': 'data/raw/charades/videos', 'mp4_files': video_count}])
```

Conclusion: this notebook only verifies that the expected local coursework inputs are present. The final reported metrics are read from saved public result files in notebooks `03`-`06`, not recomputed here.
