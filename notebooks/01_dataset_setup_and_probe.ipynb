# 01. Dataset Setup and Probe

Цель ноутбука: показать, как подготовлен датасет для курсовой. Здесь нет тяжелых вычислений: только проверка локальных файлов и короткий обзор структуры.


Основной датасет экспериментов: **Charades-STA**. QVHighlights оставлен как preliminary-направление: код и ранние результаты сохранены, но YouTube-зависимые raw videos оказались менее воспроизводимыми для курсовой.


```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path.cwd()
if not (ROOT / 'results').exists():
    ROOT = ROOT.parent

def read_json(path):
    with open(ROOT / path, 'r', encoding='utf-8') as f:
        return json.load(f)

def metrics_table(rows):
    return pd.DataFrame(rows)

paths = [
    'data/charades_sta/charades_sta_train.txt',
    'data/charades_sta/charades_sta_test.txt',
    'data/raw/charades/videos',
    'docs/data_setup.md',
    'data/qvhighlights',
]
status = []
for rel in paths:
    p = ROOT / rel
    status.append({'path': rel, 'exists': p.exists(), 'type': 'dir' if p.is_dir() else 'file'})
pd.DataFrame(status)
```

```python
def count_lines(rel):
    p = ROOT / rel
    if not p.exists():
        return None
    with p.open('r', encoding='utf-8') as f:
        return sum(1 for _ in f)

pd.DataFrame([
    {'split': 'train', 'path': 'data/charades_sta/charades_sta_train.txt', 'rows': count_lines('data/charades_sta/charades_sta_train.txt')},
    {'split': 'test', 'path': 'data/charades_sta/charades_sta_test.txt', 'rows': count_lines('data/charades_sta/charades_sta_test.txt')},
])
```

```python
video_dir = ROOT / 'data/raw/charades/videos'
video_count = len(list(video_dir.glob('*.mp4'))) if video_dir.exists() else 0
pd.DataFrame([{'video_dir': 'data/raw/charades/videos', 'mp4_files': video_count}])
```
